## Randomized Benchmarking

```python
#Import general libraries (needed for functions)
import numpy as np
import matplotlib.pyplot as plt
from IPython import display

#Import the RB Functions
import qiskit.ignis.verification.randomized_benchmarking as rb

#Import Qiskit classes 
import qiskit
from qiskit.providers.aer.noise import NoiseModel
from qiskit.providers.aer.noise.errors.standard_errors import depolarizing_error, thermal_relaxation_error
```

```python
#Generate RB circuits (2Q RB)

#number of qubits
nQ=2
rb_opts = {}
#Number of Cliffords in the sequence
rb_opts['length_vector'] = [1] +[5*i for i in range(1, 25)]
#Number of seeds (random sequences)
rb_opts['nseeds'] =1
#Default pattern
rb_opts['rb_pattern'] = [[0, 1]]

rb_circs, xdata = rb.randomized_benchmarking_seq(**rb_opts)
```

```python
# converting everything to u3

from qiskit.compiler import transpile
basis_gates =  ['id', 'u3', 'cx']

rb_u3_circs = []

for i in range(len(rb_circs)): 
    seed_circs = []
    for j in range(len(rb_circs[i])): 
        transpiled_circ = transpile(rb_circs[i][j], basis_gates = basis_gates)
        seed_circs.append(transpiled_circ)
    rb_u3_circs.append(seed_circs)
```

```python
#since only one seed
rb_u3_circs = rb_u3_circs[0]
```

```python
def removeMeasurement(circ):
    '''
    Removes the measurement gates from the circuit
    '''
    newqasm_str=""
    circ_str=circ.qasm()

    for line in iter(circ_str.splitlines()):

        if line.startswith('measure'):
            continue;
        else:
            newqasm_str+= line
    circo=QuantumCircuit().from_qasm_str(newqasm_str)
    return circo

new_circ = []
for circ in rb_u3_circs:
    n_circ = removeMeasurement(circ)
    new_circ.append(n_circ)
rb_u3_circs = new_circ
```
